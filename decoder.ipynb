```python
import os
import copy
from itertools import count
import json
import pickle
import numpy as np
import csv

import matplotlib.pyplot as plt
```

```python
# open json file
path = './csv-list-data/'
env_file = os.path.join(path, 'CaveCrawler-v0.csv')

env_csv = []
with open(env_file) as f:
    env = csv.reader(f)
    for row in env:
        env_csv.append(row)

text_data = np.flipud(env_csv)
```

```python
env_dict = {}
```

```python
grid_width,grid_height = len(env_csv[0]),len(env_csv)-5
grid_width, grid_height
```

```python
env_dict["grid_width"] = grid_width
env_dict["grid_height"] = grid_height
env_dict
```

```python
def rinsetu(A,i,j):
    rinsetu = []
    X = len(A[0])
    if i > 0:
        if A[i-1][j] == 'H' or A[i-1][j] == 'S':
            rinsetu.append((i-1)*X + j)
    if i < len(A)-1:
        if A[i+1][j] == 'H' or A[i+1][j] == 'S':
            rinsetu.append((i+1)*X + j)
    if j > 0:
        if A[i][j-1] == 'H' or A[i][j-1] == 'S':
            rinsetu.append(i*X + j-1)
    if j < X-1:
        if A[i][j+1] == 'H' or A[i][j+1] == 'S':
            rinsetu.append(i*X + j+1)
    return rinsetu
```

```python
object_dict = {}
terrain = {}
indice_list = []
type_list = []
neighbors = {}
for i in range(len(text_data)):
    text1 = list(text_data[i])
    for j,t in enumerate(text1):
        if t == 'H':
            indice_list.append(j+grid_width*i)
            type_list.append(5)
            neighbors[str(j+grid_width*i)] = rinsetu(text_data,i,j)
        if t == 'S':
            indice_list.append(j+grid_width*i)
            type_list.append(2)
            neighbors[str(j+grid_width*i)] = rinsetu(text_data,i,j)

terrain['indice'] = indice_list
terrain['type'] = type_list
terrain['neighbors'] = neighbors
object_dict['terrain'] = terrain
```

```python
env_dict['object'] = object_dict
```

```python
with open('./test-json-data/test.json', 'w') as f:
    json.dump(env_dict, f)
```
